```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
from nltk.corpus import stopwords
import sys, os
os.chdir(sys.path[0])

# Read CSV file into a pandas DataFrame
df2 = pd.read_csv('updated_neg.csv')

# Extract text from a specific column
updated_neg = df2['Text'].to_string(index=False)

# Save the text to a text file
with open('updated_neg.txt', 'w', encoding='utf-8') as file:
    file.write(updated_neg)
    
text1 = open('updated_neg.txt', mode = 'r', encoding = 'utf-8').read()
stopwords = set(stopwords.words('english'))

wc = WordCloud(
        background_color = 'white',
        stopwords= stopwords,
        height = 600,
        width = 400)

wc.generate(text1)
wc.to_file('updated_neg.png')
```
